```python
class MyHashMap:

    def __init__(self):
        self.arr = [-1] * 200
        print(self.arr)

    def put(self, key, value):
        self.arr[key] = value
        print(self.arr)

    def get(self, key):
        return self.arr[key]

    def remove(self, key):
        self.arr[key] = -1
```

```python
# Your MyHashMap object will be instantiated and called as such:
obj = MyHashMap()
obj.put(1,"mona")
obj.put(1,"mo")
param_2 = obj.get(1)
obj.remove(1)
```
